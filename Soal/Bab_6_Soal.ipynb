```python
# Problem 6.1 - Mudah
def rumus(a,b,c) :
    return 2*a**3 + 5*b**2 - c

a = int(input("Masukkan a: "))
b = int(input("Masukkan b: "))
c = int(input("Masukkan c: "))
print(f"f{a,b,c} =", rumus(a,b,c))
```

```python
# Problem 6.2 - Mudah
def fungsi_f(a) :
    return 2/3*a**2 + 5*a + 7

def fungsi_g(a) :
    return a**0.5 + 3/a

x = int(input("Masukkan bilangan x: "))
print(f"f(f(g({x}))) =", "{:.4f}".format(fungsi_f((fungsi_f(fungsi_g(x))))))
print(f"f(g(f({x}))) =", "{:.4f}".format(fungsi_f(fungsi_g(fungsi_f(x)))))
print(f"g(f(f({x}))) =", "{:.4f}".format(fungsi_g(fungsi_f(fungsi_f(x)))))
```

```python
# Problem 6.3 - Sedang
def fpb(a,b) :
    if a > b :
        b, a = a, b
    for i in range(a,0,-1) :
        if a % i == 0 and b % i == 0 :
            return i

def kpk(a,b) :
    return a * b / fpb(a,b)

a = int(input("Masukkan bilangan A: "))
b = int(input("Masukkan bilangan B: "))
c = int(input("Masukkan bilangan C: "))
print(f"KPK dari {a}, {b}, dan {c} adalah", int(kpk(a,kpk(b,c))))
```

```python
# Problem 6.4 - Sedang
def cekPrima(a) :
    for i in range(2,a) :
        if a % i == 0 :
            return False
    return True

a = int(input("Masukkan bilangan A: "))
b = int(input("Masukkan bilangan B: "))

print("Pasangan bilangan komposit:")
for i in range(a,b+1) :
    for j in range(i, b+1) :
        if i != j and not(cekPrima(i)) and not(cekPrima(j)) and not(cekPrima(i+j)) :
            print(i,j)
```

```python
# Problem 6.5 - Sedang
def sempurna(x) :
    res = 0
    for i in range(1,x) :
        if x % i == 0 :
            res += i

    return res == x

a = int(input("Masukkan bilangan A: "))
b = int(input("Masukkan bilangan B: "))

print("Bilangan sempurna:")
for i in range(a,b+1) :
    if sempurna(i) :
        print(i)
```

```python
# Problem 6.6 - Sedang
def destohex(desi) :
    if int(desi) < 10 :
        return int(desi)
    elif int(desi) >= 10 and int(desi) <= 15 :
        return chr(55+int(desi))
def hextodes(hexa) :
    if hexa.isdigit() :
        return int(hexa)
    else :
        return ord(hexa) - 55


a = input("Masukkan bilangan A: ")
b = input("Masukkan bilangan B: ")

resA = hextodes(a[0]) * 16 + hextodes(a[1])
resB = hextodes(b[0]) * 16 + hextodes(b[1])

hex1 = destohex((resA + resB) // 16)
hex2 = destohex((resA + resB) % 16)

print(f"{a} + {b} = {hex1}{hex2}")
```

```python
# Problem 6.7 - Sulit
def cek(arr, nama, n) :
    c = 0
    for i in range(n) :
        if arr[i] == nama :
            c += i
    return c

n = int(input("Masukkan jumlah barang yang dibeli: "))
arr = ["0" for i in range(n)]

for i in range(n) :
    arr[i] = input(f"Masukkan barang ke-{i+1}: ")

keranjang = True
if cek(arr, 'Sabun', n) > 3 :
    keranjang = False
if cek(arr, 'Sabun', n) > 2 :
    if cek(arr, 'HandSanitizer', n) > 2 :
        keranjang = False
elif cek(arr, 'HandSanitizer', n) > 3 :
    keranjang = False

if cek(arr, 'Faceshield', n) > 0 :
    if cek(arr, 'Mask', n) > 2 :
        keranjang = False
elif cek(arr, 'Mask', n) > 4 :
    keranjang = False

if cek(arr, 'Mask', n) > 0 :
    if cek(arr, 'Faceshield', n) > 2 :
        keranjang = False
elif cek(arr, 'Faceshield', n) > 4 :
    keranjang = False

if keranjang :
    print("Keranjang belanja diperbolehkan")
else :
    print("Keranjang belanja tidak diperbolehkan")
```

```python
# Problem 6.8 - Sulit
def balik(arr, a, b) :
    for i in range(a, b+1) :
        for j in range(a-1, b-(i-a)-1) :
            arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
    
def tambah(arr, a, b, c) :
    for i in range(a-1, b) :
        arr[i] += c
    return arr
        
n = int(input("Masukkan N: "))
arr = [0 for i in range(n)]

for i in range(n) :
    arr[i] = int(input(f"Masukkan bilangan ke-{i+1}: "))

m = int(input("Masukkan M: "))
for i in range(m) :
    operasi = input(f"Operasi ke-{i+1}: ")
    if operasi == 'balik' :
        a = int(input("Interval kiri: "))
        b = int(input("Interval kanan: "))
        arr = balik(arr, a, b)
    else :
        a = int(input("Interval kiri: "))
        b = int(input("Interval kanan: "))
        c = int(input("Nilai Tambah: "))
        arr = tambah(arr, a, b, c)

    print("Nilai Akhir :", end="")
    for i in range(n) :
        print(arr[i], end=" ")
```

```python
def input_data(formatCode, N, M) :
    if formatCode == 0 :
        data = [[input(f"Masukkan entry baris ke-{i+1} kolom ke-{j+1}")
            for i in range(M)]
            for j in range(N)]
    else :
        print("Input:")
        data = [input().split(",") for i in range(N)]
    return data

def padded_print(string, length): 
    print(string, end="")
    for i in range(len(string), length) :
        print(" ", end="")
def horizontal_print(length) :
    for i in range(length) :
        print("-", end="")
    print()

def output_data(data, N, M) :
    max_length = [0 for j in range(M)]
    for i in range(N) :
        for j in range(M) :
            max_length[j] = max(max_length[j], len(data[i][j]))
    total_length = 1
    for j in range(M) :
        total_length += max_length[j] + 1

    for i in range(N) :
        if i <= 1 :
            print(total_length*"-")
        for j in range (M) :
            print("|", end='')
            padded_print(data[i][j], max_length[j])
        print("|")
    horizontal_print(total_length)

N = int(input("Masukkan N: "))
M = int(input("Masukkan M: "))
data = input_data(0, N, M)
print("Output")
output_data(data, N, M)
```
