<h3>Problem 6.1 - Mudah</h3>
<p>Tuan Wang ingin memiliki sebuah rumus <i>f</i>. Fungsi ini didefinisikan sebagai</p>
<p align="center"><i>f(a,b,c) = 2a<sup>3</sup> + 5b<sup>2</sup> - c</i></p>
<p>Ia akan memberikan Anda nilai <i>a, b</i> dan <i>c</i>, tentukan nilai <i>f(a,b,c)</i></p>

<h4>Contoh</h4>
<p>Masukkan a: <u>10</u><br>
Masukkan b: <u>3</u><br>
Masukkan c: <u>5</u><br>
f(10, 3, 5) = 2040</p>

```python
# Problem 6.1 - Mudah
def rumus(a,b,c) :
    return 2*a**3 + 5*b**2 - c

a = int(input("Masukkan a: "))
b = int(input("Masukkan b: "))
c = int(input("Masukkan c: "))
print(f"f{a,b,c} =", rumus(a,b,c))
```

<h3>Problem 6.3 - Sedang</h3>
<p>Tuan Wang tahu mengenai Faktor Persekutuan Terbesar (FPB) dan Kelipatan Persekutuan Terkecil (KPK). Namun, ia bingung bagaimana cara membuat program yang menghitung KPK dari 3 bilangan. Yang ia tahu, KPK dari dua bilangan dapat dihitung dengan :</p>

<p align="center"><i>a x b<br>
<span style="text-decoration:overline;">FPB(a,b)</span></i></p>

<p>Pada dasarnya, KPK dari a, b, dan c adalah KPK dari a dan P, apabila P adalah KPK dari b dan c. Buatkanlah program untuk menghitung KPK dari tiga bilangan.</p>

<h4>Contoh</h4>
<p>Masukkan bilangan A: <u>8</u><br>
Masukkan bilangan B: <u>6</u><br>
Masukkan bilangan C: <u>14</u><br>
KPK dari 8, 6, dan 14 adalah 168<p>

```python
# Problem 6.3 - Sedang
def fpb(a,b) :
    if a > b :
        b, a = a, b
    for i in range(a,0,-1) :
        if a % i == 0 and b % i == 0 :
            return i

def kpk(a,b) :
    return a * b / fpb(a,b)

a = int(input("Masukkan bilangan A: "))
b = int(input("Masukkan bilangan B: "))
c = int(input("Masukkan bilangan C: "))
print(f"KPK dari {a}, {b}, dan {c} adalah", int(kpk(a,kpk(b,c))))
```

<ul>
    <li>6 dan 9 merupakan bilangan komposit dan 6 + 9 = 15 juga merupakan bilangan komposit, maka 6 dan 9 adalah <i> pasangan bilangan komposit</i></li>

<h4>Contoh</h4>

4 6<br>4 8<br>4 10<br>6 8<br>6 9<br>6 10<br>8 10</p>

<h3>Problem 6.5 - Sedang</h3>
<p>Tuan Wang sedang menguji kemampuannya. Ia ingin mengatahui bilangan sempurna diantara dua bilangan. Bilangan sempurna adalah bilangan bulat positif yang merupakan penjumlahan dari faktor-faktornya kecuali bilangan itu sendiri. Perhatikan contoh berikut:</p>
<ul>
    <li>6 Memiliki faktor 1, 2, 3, 6. Perhatikan bahwa 6 = 1 + 2 + 3, maka 6 merupakan bilangan sempurna</li>
    <li>28 merupakan bilangan sempurna karena 28 = 1 + 2 + 4 + 7 + 14</li>
</ul>

<h4>Contoh</h4>
<p>Masukkan bilangan A: <u>1</u><br>
Masukkan bilangan B: <u>50</u><br>
Bilangan sempurna:<br>
6<br>28</p>

```python
# Problem 6.5 - Sedang
def sempurna(x) :
    res = 0
    for i in range(1,x) :
        if x % i == 0 :
            res += i

    return res == x

a = int(input("Masukkan bilangan A: "))
b = int(input("Masukkan bilangan B: "))

print("Bilangan sempurna:")
for i in range(a,b+1) :
    if sempurna(i) :
        print(i)
```

<h3>Problem 6.6 - Sulit</h3>
<p>Tuan Wang baru saja belajar mengenai heksadesimal. Heksadesimal adalah sistem bilangan yang direpresentasikan dengan periode 16. Sebagai perbandingan, desimal adalah sistem bilangan yang direpresentasikan dengan periode 10. Perhatikan tabel translasi desimal - heksadesimal berikut:</p>

|Desimal|Heksadesimal|
|-------|------------|
|0      |00          |
|1      |01          |
|2      |02          |
|3      |03          |
|...    |...         |
|31     |1F          |
|32     |20          |
|33     |21          |
|...    |...         |
|255    |FF          |

<p>Tugas Anda sekarang adalah membuat program yang menerima dua heksadesimal dengan panjang <b>tepat</b> dua digit, melakukan penjumlahan, dan menuliskan hasilnya dalam dua digit angka heksadesimal. DIjamin input dan output bilangan dapat direpresentasikan tidak lebih dari dua digit heksadesimal.</p>

<h4>Contoh</h4>
<p>Masukkan bilangan A: <u>01</u><br>
Masukkan bilangan B: <u>29</u><br>
01 + 29 = 2A</p>

```python
# Problem 6.6 - Sedang
def destohex(desi) :
    if int(desi) < 10 :
        return int(desi)
    elif int(desi) >= 10 and int(desi) <= 15 :
        return chr(55+int(desi))
def hextodes(hexa) :
    if hexa.isdigit() :
        return int(hexa)
    else :
        return ord(hexa) - 55


a = input("Masukkan bilangan A: ")
b = input("Masukkan bilangan B: ")

resA = hextodes(a[0]) * 16 + hextodes(a[1])
resB = hextodes(b[0]) * 16 + hextodes(b[1])

hex1 = destohex((resA + resB) // 16)
hex2 = destohex((resA + resB) % 16)

print(f"{a} + {b} = {hex1}{hex2}")
```

<p>Program akan menerima masukan jumlah item yan gakan dibeli serta item apa saja yang akan dibeli. Program kemudian mencetak apakah keranjang belanja tersebut diperbolehkan sesuai aturan diatas.</p>

<h4>Contoh 1</h4>
<p>Masukkan jumlah barang yang dibeli: <u>5</u><br>

<h4>Contoh 2</h4>
<p>Masukkan jumlah barang yang dibeli: <u>6</u><br>

Masukkan barang ke-3: <u>Sabun</u><br>
Masukkan barang ke-4: <u>Sabun</u><br>
Masukkan barang ke-5: <u>Sabun</u><br>
Masukkan barang ke-6: <u>FaceShield</u><br>
Keranjang belanja diperbolehkan.</p>

```python
# Problem 6.7 - Sulit
def cek(arr, nama, n) :
    c = 0
    for i in range(n) :
        if arr[i] == nama :
            c += i
    return c

n = int(input("Masukkan jumlah barang yang dibeli: "))
arr = ["0" for i in range(n)]

for i in range(n) :
    arr[i] = input(f"Masukkan barang ke-{i+1}: ")

keranjang = True
if cek(arr, 'Sabun', n) > 3 :
    keranjang = False
if cek(arr, 'Sabun', n) > 2 :
    if cek(arr, 'HandSanitizer', n) > 2 :
        keranjang = False
elif cek(arr, 'HandSanitizer', n) > 3 :
    keranjang = False

if cek(arr, 'Faceshield', n) > 0 :
    if cek(arr, 'Mask', n) > 2 :
        keranjang = False
elif cek(arr, 'Mask', n) > 4 :
    keranjang = False

if cek(arr, 'Mask', n) > 0 :
    if cek(arr, 'Faceshield', n) > 2 :
        keranjang = False
elif cek(arr, 'Faceshield', n) > 4 :
    keranjang = False

if keranjang :
    print("Keranjang belanja diperbolehkan")
else :
    print("Keranjang belanja tidak diperbolehkan")
```

<h3>Problem 6.8 - Sulit</h3>
<p>Tuan Mor suka bermain dengan angka. Karena beliau sudah pandai berhitung, beliau ingin menantang dirinya untuk bermain dan berhitung tidak hanya dengan satu atau dua angka, melainkan sekumpulan angka. Tuan Mor awalnya akan menuliskan barisan n buah angka, sebut berindex 1 hingga n. Beliau kemudian meminta temannya, Tuan Rom untuk memberikan m buah operasi. Operasi yang diberikan Tuan Rom hanya 2 jenis dengan format:</p>

<ul>
    <li><i>tambah i j k, 1 ≤ i ≤ j < n,</i> operasi ini menambahkan bilangan dengan indeks antara i sampai j (inklusif) dengan bilangan bulat k</li>
    <li><i>balik i j, 1 ≤ i ≤ j n,</i> operasi ini membalik urutan subbarisan bilangan dengan indeks antara i sampai j (inklusif)</li>
</ul>

<p>Tuan Mor pun kesusahan adan karena dia mudah menyerah (jangan ditiru lol), Anda pun diminta untuk membuat prosedur program dengan spesifikasi input sesuai format yang sudah dijabarkan diatas. Buatlah prosedur tambah dan balik untuk membantu main program Anda. Outputkan hasil akhir dari barisan bilangan setelah dilakukan setiap operasi secara berurutan. Anda boleh mengimplrementasikan salah satu dari 2 cara program contoh menerima output berikut:</p>

<h4>Contoh 1</h4>
<p>Masukkan N: <u>4</u><br>
Masukkan bilangan ke-1: <u>2</u><br>
Masukkan bilangan ke-2: <u>-1</u><br>
Masukkan bilangan ke-3: <u>0</u><br>
Masukkan bilangan ke-4: <u>2</u></p>

<p>Masukkan M: <u>2</u><br>
Operasi ke-1: <u>balik</u><br>
Interval kiri: <u>2</u><br>
Interval kanan: <u>4</u><br>
Operasi ke-2: <u>tambah</u><br>
Interval kiri: <u>1</u><br>
Interval kanan: <u>3</u><br>
Nilai tambah: <u>-1</u></p>

<p>Hasil akhir: 1 1 -1 -1</p>

<h4>Contoh 2</h4>
<p>Masukkan N: <u>5</u><br>
Masukkan array: <u>2 -1 0 2 -3</u></p>

<p>Masukkan M: <u>2</u><br>
Operasi ke-1: <u>tambah 1 5 3</u><br>
Operasi ke-2: <u>balik 3 3</u></p>

<p>Hasil akhir: 5 4 3 5 0</p>





```python
# Problem 6.8 - Sulit
def balik(arr, a, b) :
    for i in range(a, b+1) :
        for j in range(a-1, b-(i-a)-1) :
            arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
    
def tambah(arr, a, b, c) :
    for i in range(a-1, b) :
        arr[i] += c
    return arr
        
n = int(input("Masukkan N: "))
arr = [0 for i in range(n)]

for i in range(n) :
    arr[i] = int(input(f"Masukkan bilangan ke-{i+1}: "))

m = int(input("Masukkan M: "))
for i in range(m) :
    operasi = input(f"Operasi ke-{i+1}: ")
    if operasi == 'balik' :
        a = int(input("Interval kiri: "))
        b = int(input("Interval kanan: "))
        arr = balik(arr, a, b)
    else :
        a = int(input("Interval kiri: "))
        b = int(input("Interval kanan: "))
        c = int(input("Nilai Tambah: "))
        arr = tambah(arr, a, b, c)

    print("Nilai Akhir :", end="")
    for i in range(n) :
        print(arr[i], end=" ")
```

<p>Tuan Far merupakan seorang data analis terkenal sekaligus tokoh idola Anda. Suatu hari anda mendapat kesempatan untuk mendapat petuah darinya. Tua nfar pun mengungkapkan bahwa rahasia sukses data analis adalah menguasai pengolahan data dalam bentuk tabel. Anda pun mendapat ilham untuk latihan mengolah tabel. <br>

Program akan menulis ulang array yang dimasukkan dengan menambahkan beberapa karakter yang diperlukan agar terlihat seperti tabel. Karakter yang digunakan ialah - untuk garis horizontal dan | untuk garis vertikal. Garis horizontal hanya dituliskan pada ujung ataas jan ujung bawah tabel, serta diantar baris ke-1 dan ke-2 (jika ada). Gunakan salah satu format input yang diberikan pada contoh berikut:</p>

<h4>Contoh 1</h4>
<p>Masukkan N: <u>3</u><br>
Masukkan M: <u>2</u><br>
Masukkan entry baris ke-1 kolom ke-1: <u>Nama Item</u><br>
Masukkan entry baris ke-1 kolom ke-2: <u>Harga</u><br>

<p>Output:<br>
------------------------<br>
|Nama Item|Harga&nbsp;&nbsp;&nbsp;&nbsp;|<br>
------------------------<br>
|Sabun&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;|1250000|<br>

```python
def input_data(formatCode, N, M) :
    if formatCode == 0 :
        data = [[input(f"Masukkan entry baris ke-{i+1} kolom ke-{j+1}")
            for i in range(M)]
            for j in range(N)]
    else :
        print("Input:")
        data = [input().split(",") for i in range(N)]
    return data

def padded_print(string, length): 
    print(string, end="")
    for i in range(len(string), length) :
        print(" ", end="")
def horizontal_print(length) :
    for i in range(length) :
        print("-", end="")
    print()

def output_data(data, N, M) :
    max_length = [0 for j in range(M)]
    for i in range(N) :
        for j in range(M) :
            max_length[j] = max(max_length[j], len(data[i][j]))
    total_length = 1
    for j in range(M) :
        total_length += max_length[j] + 1

    for i in range(N) :
        if i <= 1 :
            print(total_length*"-")
        for j in range (M) :
            print("|", end='')
            padded_print(data[i][j], max_length[j])
        print("|")
    horizontal_print(total_length)

N = int(input("Masukkan N: "))
M = int(input("Masukkan M: "))
data = input_data(0, N, M)
print("Output")
output_data(data, N, M)
```
