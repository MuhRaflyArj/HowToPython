```python
# 1. Banyaknya data
len(df)
```

```python
# 2. Pie chart banyaknya mahasiswa tiap fakultas
df['fakultas'].value_counts().plot(kind='pie')
```

```python
# 3. Pie chart banyaknya mahasiswa FTSL berdasarkan tingkat
df.loc[df['fakultas'] == 'FTSL']['tingkat'].value_counts().plot(kind='pie')
```

```python
# 4. Histogram dengan IP sebagai sumbu x dan jumlah mahasiswa tiap fakultas sebagai sumbu y, gunakan pertambahan 0.5
df['ip'].plot(kind='hist', bins=[0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4], rwidth = 0.9)
```

```python
# 5. Berdasarkan plot sebelumnya, range IP mana yang paling banyak diperoleh?

# Range 3 - 3.5
```

```python
# 6. Stacked bar chart dengan tingkat sebagai sumbu x dan jumlah mahasiswa setiap fakultas sebagai sumbu y
df.groupby(['tingkat', 'fakultas'])['nama'].size().unstack().plot(kind='bar', stacked=True)
```

```python
# 7. Berdasarkan plot sebelumnya, angkatan STEI yang jumlah mahasiswanya paling sedikit ?

# Jumlah mahasiswa STEI paling sedikit pada angkatan 2018
df.loc[df['fakultas'] == 'STEI'].groupby(['tingkat'])['fakultas'].value_counts().sort_index(ascending=False).plot(kind='barh')
```

```python
# 8. Line chart dengan tingkat sebagai sumbu x dan jumlah mahasiswa sumbu y
df['tingkat'].value_counts().sort_index().plot(kind='line', y='counts', x='tingkat')
```

```python
# 9. Line chart seperti soal sebelumnya, namun terdapat 4 garis, masing-masing untuk tiap fakultas
df.groupby(['tingkat', 'fakultas'])['nama'].size().unstack().plot(kind='line')
```

```python
# 10. Berdasarkan plot sebelumnya, manakah yang jumlah mahasiswanya terus bertambah ?

# Berdasarkan grafik, dapat dilihat bahwa fakultas yang mahasiswanya terus bertambah adalah SITHR
df.loc[df['fakultas'] == 'SITHR'].groupby(['tingkat'])['fakultas'].value_counts().sort_index(ascending=False).plot(kind='barh')
```

```python
# 11. Berdasarkan plot sebelumnya, manakah yang jumlah mahasiswanya terus berkurang ?

# Berdasarkan grafik, dapat dilihat bahwa fakultas yang mahasiswanya terus berkurang adalah STEI
df.loc[df['fakultas'] == 'STEI'].groupby(['tingkat'])['fakultas'].value_counts().sort_index(ascending=False).plot(kind='barh')
```
